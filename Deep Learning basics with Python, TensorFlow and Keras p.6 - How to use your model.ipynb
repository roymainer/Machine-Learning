## Loading a model:

```python
import cv2
import tensorflow as tf

CATEGORIES = ["Dog", "Cat"]  # re-convert from bool (0/1) to Dog/Cat


def prepare(filepath):
    # prepare the image
    IMG_SIZE = 50
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

model = tf.keras.models.load_model('dogs_cats_detector.model')  # load the model

```

## Predicitons:

```python
prediction = model.predict([prepare('dog.jpeg')])  # always pass a list
print(CATEGORIES[int(prediction[0][0])])
```

```python
prediction = model.predict([prepare('cat.jpg')])  # always pass a list
print(CATEGORIES[int(prediction[0][0])])
```
